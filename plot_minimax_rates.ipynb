```python
import numpy as np
import matplotlib.pyplot as plt
from cProfile import label
from scipy.optimize import curve_fit
```

# Minimax rates: torus

```python
# load results
OT_loss = np.array(np.load('results/torus/OT_losses_avg_tau0.npy'))
```

```python
# p = 1
def objective_func(x,a,b):
    return a * (x) ** (-b) * np.log2(x)
popt, _ = curve_fit(objective_func,np.arange(10,510,10), OT_loss[:,0])
a,b = popt
plt.loglog(np.arange(10,510,10),OT_loss[:,0],color='tab:purple',label='$p=1$') 
plt.loglog(np.arange(10,510,10),objective_func(np.arange(10,510,10),a,b),'--',color='tab:purple',label='fit: $%.2e \cdot N^{-%.2f} \cdot \log_2{N}$' % (a,b))
# p = 2
def objective_func(x,a,b):
    return a * (x) ** (-b)
popt, _ = curve_fit(objective_func,np.arange(10,510,10), OT_loss[:,2])
a,b = popt
plt.loglog(np.arange(10,510,10),OT_loss[:,2],color='tab:blue',label='$p=2$') 
plt.loglog(np.arange(10,510,10),objective_func(np.arange(10,510,10),a,b),'--',color='tab:blue',label='fit: $%.2e \cdot N^{-%.2f}$' % (a,b))
# p = 3
popt, _ = curve_fit(objective_func,np.arange(10,510,10), OT_loss[:,4])
a,b = popt
plt.loglog(np.arange(10,510,10),OT_loss[:,4],color='tab:green',label='$p=3$') 
plt.loglog(np.arange(10,510,10),objective_func(np.arange(10,510,10),a,b),'--',color='tab:green',label='fit: $%.2e \cdot N^{-%.2f}$' % (a,b))
# p = 4
popt, _ = curve_fit(objective_func,np.arange(10,510,10), OT_loss[:,6])
a,b = popt
plt.loglog(np.arange(10,510,10),OT_loss[:,6],color='tab:orange',label='$p=4$') 
plt.loglog(np.arange(10,510,10),objective_func(np.arange(10,510,10),a,b),'--',color='tab:orange',label='fit: $%.2e \cdot N^{-%.2f}$' % (a,b))

plt.xlabel(r"$N$")
plt.ylabel(r"$\mathrm{OT}_p^p(\hat{\mu}_{\mathrm{H}},\bar{\mu}_{M})$") # M = 10000
plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0,fancybox=True,shadow=True,ncol=1)
plt.minorticks_on()
plt.grid(True,which='minor')
plt.show()
```
